## 4. Pré-processamento e Limpeza de Dados
- **Tratamento de Outliers:**

```python
df.drop(df[df['genre'] == 'set()'].index, inplace=True)
```

- **Correção de Tipos de Dados:**

```python
df['genre_adjusted'] = df['genre'].str.split(", ")
df = df.explode('genre_adjusted')
df['genre_adjusted'].unique()
```

---
## 5. Engenharia de Atributos (Feature Engineering)
- **Criação de novas variáveis:**

```python
df2015 = df[df['year'] > 2015]
```

- **Escalonamento de Variáveis (Scaling/Normalization):**

```python
df['duration_mn'] = (df['duration_ms'] / 60000).round(2)
```

### Análise: As músicas de qual ano são as de mais sucesso?

```python
df.groupby("year")['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com mais músicas escutadas durante esse intervalo de anos?

```python
df.groupby('artist')['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com mais músicas escutadas durante os anos de 2015 à 2019?

```python
df2015.groupby('artist')['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o gênero musical mais popular?

```python
df.groupby('genre_adjusted')['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com músicas mais "positivas"? (de acordo com o nível de valência)

```python
df.groupby('artist')['valence'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com músicas mais "positivas" entre os anos de 2015 à 2019? (de acordo com o nível de valência)

```python
df2015.groupby('artist')['valence'].mean().sort_values(ascending=False)
```

### Análise: Quais as músicas ao vivo mais escutadas?

### Análise: Qual o ano com mais músicas "negativas" lançadas? (de acordo com o nível de valência)

### Análise: Qual artista possui as músicas com maior medida de "Speechiness"?
