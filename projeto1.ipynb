## Instalação do Pandas e leitura da base de dados

```python
%pip install pandas
import pandas as pd
```

```python
df = pd.read_csv("songs_normalize.csv", index_col=False)
df.head(20)
```

### Informações

##### artist = nome do artista
##### song = nome da música
##### duration_ms = duração em milissegundos
##### explicit = se a letra é explícita ou não
##### year = ano de lançamento da faixa
##### popularity = popularidade da faixa
##### danceability = o quão adequada essa música é para dançar
##### energy = medida perceptiva de intensidade e atividade
##### loudness = notação padrão de classe de altura
##### mode = modalidade melódica da faixa
##### speechiness = a característica "speechiness" (fala) detecta a presença de palavras faladas em uma faixa (usada para medir e identificar podcasts, audiobook, talk show, etc)
##### acousticness = acústica da música
##### instrumentalness = prevê se a faixa não contém vocais
##### liveness = verifica se há a presença de público na música
##### valence = uma medida que descreve a positividade musical transmitida pela faixa
##### tempo = o andamento geral estimado de uma faixa em batidas por minuto (BPM)
##### genre = gênero musical

### Verificando colunas

```python
print(df.columns)
```

### Conversão dos milissegundos para minutos (coluna "duration_ms")

```python
df['duration_mn'] = (df['duration_ms'] / 60000).round(2)
df
```

### Análise: As músicas de qual ano são as de mais sucesso?

```python
df.groupby("year")['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com mais músicas escutadas durante esse intervalo de anos?

```python
df.groupby('artist')['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com mais músicas escutadas durante os anos de 2015 à 2019?

```python
df2015 = df[df['year'] > 2015]
df2015
```

```python
df2015.groupby('artist')['popularity'].mean().sort_values(ascending=False)
```

### Ajuste da coluna "genre"

```python
df.drop(df[df['genre'] == 'set()'].index, inplace=True)
df['genre_adjusted'] = df['genre'].str.split(", ")
```

### Análise: Qual o gênero musical mais popular?

```python
df = df.explode('genre_adjusted')
df['genre_adjusted'].unique()
df.groupby('genre_adjusted')['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com músicas mais "positivas"? (de acordo com o nível de valência)

```python
df.groupby('artist')['valence'].mean().sort_values(ascending=False)
```

### Análise: Qual o artista com músicas mais "positivas" entre os anos de 2015 à 2019? (de acordo com o nível de valência)

```python
df2015.groupby('artist')['valence'].mean().sort_values(ascending=False)
```

### Análise: Quais as músicas ao vivo mais escutadas?

```python
df.groupby(['song', 'liveness'])['popularity'].mean().sort_values(ascending=False)
```

### Análise: Qual o ano com mais músicas "negativas" lançadas? (de acordo com o nível de valência)

```python
df.groupby('year')['valence'].mean().sort_values(ascending=True)
```

### Análise: Qual artista possui as músicas com maior medida de "Speechiness"? 

```python
df.groupby('artist')['speechiness'].mean().sort_values(ascending=False)
```

### Análise: Qual artista possui mais faixas com letras explícitas?

```python
df.groupby('artist')['explicit'].count().sort_values(ascending=False)
```

### Análise: Qual gênero musical possui maior taxa de dançabilidade?

```python
df = df.explode('genre_adjusted')
df['genre_adjusted'].unique()
df.groupby('genre_adjusted')['danceability'].mean().sort_values(ascending=False)
```

### Análise: Quais foram as músicas mais escutadas de cada ano? (de 2015 à 2019)

```python
df2015.groupby('song')['popularity'].mean().sort_values(ascending=False)
```
